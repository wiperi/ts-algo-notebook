自顶向下（递归）

```py
# 自顶向下递归的动态规划

def dp(状态1, 状态2, ...):

    for 选择 in 所有可能的选择:
        # 此时的状态已经因为做了选择而改变
        result = min(result, dp(状态1, 状态2, ...))

    return result
```

顶底向上（迭代）

```py
# 自底向上迭代的动态规划

# 初始化 base case
dp[0][0][...] = base case

# 进行状态转移
for 状态1 in 状态1的所有取值：
    for 状态2 in 状态2的所有取值：
        for ...
            dp[状态1][状态2][...] = min(选择1，选择2...)
```
