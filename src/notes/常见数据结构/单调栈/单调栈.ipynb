维护 arr[i..j] ，按照降序维护任何 >= arr[j] 的元素

维护 arr[i..j] ，按照升序维护任何 <= arr[j] 的元素

当遍历到 i 的时候，单调栈中只会包含 》= nums[i] 的元素

```python
# 单调栈模板代码

import numpy as np
from typing import List


# 1. 左侧第一个大元素
def prev_greater_element(nums: List[int]) -> List[int]:
    n = len(nums)
    result = [-1] * n  # 默认为-1，表示不存在这样的元素
    stack = []  # 单调递减栈

    for i in range(n):
        # 当前元素比栈顶元素大，栈顶元素就是当前元素左侧第一个较大的元素
        while stack and nums[i] <= nums[stack[-1]]:
            stack.pop()

        if stack:
            result[i] = stack[-1]  # 记录位置

        stack.append(i)

    return result


# 2. 左侧第一个小元素
def prev_smaller_element(nums: List[int]) -> List[int]:
    n = len(nums)
    result = [-1] * n  # 默认为-1，表示不存在这样的元素
    stack = []  # 单调递增栈

    for i in range(n):
        # 当前元素比栈顶元素小，栈顶元素就是当前元素左侧第一个较小的元素
        while stack and nums[i] >= nums[stack[-1]]:
            stack.pop()

        if stack:
            result[i] = stack[-1]  # 记录位置

        stack.append(i)

    return result


# 3. 右侧第一个大元素
def next_greater_element(nums: List[int]) -> List[int]:
    n = len(nums)
    result = [-1] * n  # 默认为-1，表示不存在这样的元素
    stack = []  # 单调递减栈

    for i in range(n - 1, -1, -1):
        # 当前元素比栈顶元素大，栈顶元素的右侧第一个较大元素就是当前元素
        while stack and nums[i] >= nums[stack[-1]]:
            stack.pop()

        if stack:
            result[i] = stack[-1]  # 记录位置

        stack.append(i)

    return result


# 4. 右侧第一个小元素
def next_smaller_element(nums: List[int]) -> List[int]:
    n = len(nums)
    result = [-1] * n  # 默认为-1，表示不存在这样的元素
    stack = []  # 单调递增栈

    for i in range(n - 1, -1, -1):
        # 当前元素比栈顶元素小，栈顶元素的右侧第一个较小元素就是当前元素
        while stack and nums[i] <= nums[stack[-1]]:
            stack.pop()

        if stack:
            result[i] = stack[-1]  # 记录位置

        stack.append(i)

    return result


# 示例
if __name__ == "__main__":
    nums = [4, 2, 3, 5, 1]
    print("原始数组:", nums)
    print("左侧第一个大元素的索引:", prev_greater_element(nums))
    print("左侧第一个小元素的索引:", prev_smaller_element(nums))
    print("右侧第一个大元素的索引:", next_greater_element(nums))
    print("右侧第一个小元素的索引:", next_smaller_element(nums))
```
